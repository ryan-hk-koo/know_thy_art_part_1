**Note** :

- We've developed an object detection system using YOLOv8m, tailored specifically for paintings (painting as the only class) 
- When a user uploads an image, our system precisely identifies and isolates the painting, seamlessly removing any backgrounds and frames
- This ensures that only the pure essence of the artwork is considered, optimizing the subsequent style classification process and eliminating potential interference from extraneous elements (i.e. backgrounds and frames)

<br>

- Import the YOLO class from the ultralytics library
- Initialize the YOLO model by loading the 'yolov8m.pt' weights

```python
from ultralytics import YOLO
model = YOLO('yolov8m.pt')
```

<br>

- Train the model using the dataset specified in 'data.yaml' located in the given path
- The training process will run for 100 epochs and stop if there's no improvement after 100 epochs (as specified by the patience)
- There are 560 Training Data and 140 Test Data
- The labels were created using [makesense](http://www.makesense.ai/)

```python
model.train(data="/content/drive/MyDrive/objd_painting_2/data.yaml", epochs=100, patience=100)
```
